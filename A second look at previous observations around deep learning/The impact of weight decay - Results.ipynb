```python
%autoreload
import sys
sys.path.insert(0, "../")

import pickle
import matplotlib
import matplotlib.pyplot as plt

from experiment_utils import plot_history
from rotation_rate_utils import plot_parameter_distances
```

```python
with open('impact_of_weight_decay.p','rb') as f:
    results = pickle.load(f)
```

```python
optimizer = 'SGD'
training_mode = 'normal'
weight_decay = False
plot_history(results[optimizer][training_mode][weight_decay]['history'])
```

```python
optimizer = 'SGD'
training_mode = 'normal'
weight_decay = True
plot_history(results[optimizer][training_mode][weight_decay]['history'])
```

```python
%matplotlib inline

matplotlib.rcParams.update({'font.size': 20})
f, (ax1, ax2) = plt.subplots(2, 1, sharex = True, figsize = (7,8))

optimizer = 'SGD'
training_mode = 'normal'
weight_decay = True
plot_parameter_distances(results[optimizer][training_mode][weight_decay]['lpdm'], ax = ax1)
ax1.set_xlabel('')
matplotlib.rcParams.update({'font.size': 15})
ax1.set_title('With weight decay')

optimizer = 'SGD'
training_mode = 'normal'
weight_decay = False
plot_parameter_distances(results[optimizer][training_mode][weight_decay]['lpdm'], ax = ax2)
matplotlib.rcParams.update({'font.size': 15})
ax2.set_title('Without weight decay')

# plt.tight_layout()
plt.savefig('figures/weight_decay.eps',format='eps', dpi=200)
# plt.savefig('figures/weight_decay.png',format='png', dpi=300)
```
